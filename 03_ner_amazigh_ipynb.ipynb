```python
import spacy

# 1. Cargamos el modelo que ya configuramos
nlp = spacy.blank("xx")

# 2. Añadimos el componente Entity Ruler al pipeline
ruler = nlp.add_pipe("entity_ruler")

# 3. Definimos nuestros "Patrones de Oro" (Tu conocimiento del terreno)
patterns = [
    {"label": "LOC", "pattern": "Tamazgha"},
    {"label": "LOC", "pattern": "Agadir"},
    {"label": "LOC", "pattern": "Nador"},
    {"label": "PER", "pattern": "Massinissa"},
    {"label": "PER", "pattern": "Dihya"},
    {"label": "ORG", "pattern": "IRCAM"}
]

ruler.add_patterns(patterns)

# 4. Probamos el modelo con una frase
text = "Dihya tzedegh g Agadir, tamane n Tamazgha."
doc = nlp(text)

# 5. Visualización profesional (DisplaCy)
from spacy import displacy
displacy.render(doc, style="ent", jupyter=True)
```

```python
# Actualizamos los patrones para incluir Tifinagh
patterns_tifinagh = [
    # CIUDADES / LUGARES
    {"label": "LOC", "pattern": "ⵍⵃⵓⵙⵉⵎⴰ"}, # Alhucima
    {"label": "LOC", "pattern": "ⴰⵣⵖⴰⵏⵖⴰⵏ"}, # Segangan
    {"label": "LOC", "pattern": "ⵉⵖⵣⴰⵔ ⵓⵎⴰⵙⵉⵏ"}, # Ighzar umassin
    {"label": "LOC", "pattern": "ⴰⵔⵔⵉⴼ"}, # Rif

    # PERSONAJES
    {"label": "PER", "pattern": "ⵎⴰⵙⵉⵏⵉⵙⴰ"} # Massinissa
]

ruler.add_patterns(patterns_tifinagh)

# Prueba bilingüe
test_text = "ⵎⴰⵙⵉⵏⵉⵙⴰ d agellid n Tamazgha. Alhucima tzeddeg g ⴰⵔⵔⵉⴼ."
doc = nlp(test_text)

displacy.render(doc, style="ent", jupyter=True)
```
